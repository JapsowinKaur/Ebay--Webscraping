#### Web Scraping Data from Ebay for sponsored and non-sponsored items

```python
# Importing necessary libraries
from bs4 import BeautifulSoup
from requests import get
from lxml import html
import nltk
import re, pprint as pp
import requests
import time
import pandas as pd
from sqlalchemy import create_engine
```

```python
# URL of the website to be scraped

# Creating two files to store the webscraped results

filename_1 = open('sponsored.txt', 'w')
filename_2 = open('non-sponsored.txt', 'w')

# Looping through the results of the first 10 pages, with 100 results/page

for i in range(1,11):
    url = "https://www.ebay.com/sch/i.html?_from=R40&_nkw=playstation+4+slim&LH_BIN=1&_ipg=100&_pgn="+str(i)
    
    res = get(url)
    data = BeautifulSoup(res.text, 'html.parser')
    items = data.find_all('a',class_ = 's-item__link')
    spon = 1
    nonspon = 1

# looping through results on the current page and segregating them into sponsored and non-sponsored using regex

    for item in items:
        url_simplify = re.sub("(.*)\\?.*", r"\1", item['href'])   
        if re.search('.*S.*P.*O.*N.*S.*O.*R.*E.*D.*',item['href']):
            filename_1.write(url_simplify+'\n')
            print(nonspon,url_simplify)
            nonspon = nonspon+1
            
        else:
            filename_2.write(url_simplify+'\n')
            print(spon, url_simplify)
            spon = spon + 1
    
# Pausing after iteration 

    time.sleep(1)
  
                             
```

```python
import os
import codecs

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

# Creating two folders in the current directory: for sponsored and non-sponsored links

createFolder('./sponsored/')
createFolder('./non-sponsored/')

# Opening the two files: sponsored.txt and non-sponsored.txt

file_spon = open('sponsored.txt', 'r')
file_non = open('non-sponsored.txt', 'r')
user_agent = {'User-agent': 'Mozilla/5.0'}

# Function to extract the ID from each link, getting the URL, parsing the URL to Python object and then writing the results
# in respective folders

def files_to_folder(item_type, file_type):
    for line in file_type:
        id_name = (line.split('/')[-1]).split('\n')[0]
        path = '{}/{}.html'.format(item_type,id_name)
        url = line.split('\n')[0]
        print(id_name,':', url)
        url_content = requests.get(url, headers = user_agent)
        soup = BeautifulSoup(url_content.text, 'html.parser')
        with codecs.open(path, 'w', encoding = 'utf8') as file:
            file.write(str(soup))
        time.sleep(1)
```

```python
# Calling the function above for sponsored links and sponsored file

files_to_folder('sponsored',file_spon)
```

```python
# Calling the function above for non-sponsored linkes and non-sponsored file

files_to_folder('non-sponsored',file_non)
```

```python
import os

# Creating lists for the required attributes

Seller_Name = []
Seller_Score = []
Item_Price = []
Num_Items_Sold = []
Best_Offer = []
Title = []
Return_Allowed = []
Shipping_Price = []
Condition = []
temp_item = []
Num_Items_Sold_items = []
eBay_item_type = []

# Opening files in sponsored folder

path =os.getcwd()+'\sponsored'
#print(path)
num = 1
count = 0

for filename in os.listdir(path):
# Opening files in folder    
    f= open(path+'\\'+filename, 'r', encoding = 'utf8') 
# Parsing the response and storing the resulting html
    data = BeautifulSoup(f.read(), 'html.parser')

# Finding all the tags and class/IDs for the atrributes
    
    seller_name = data.find_all('span',class_ = 'mbg-nw')


# Getting seller name information
    seller_count = 0
    seller_name_items = data.find_all('span',class_ = 'mbg-nw')
    for item in seller_name_items: 
        seller_count = seller_count + 1
        if seller_count < 2:
            Seller_Name.append(item.text) 


# Getting seller score information
    score_count = 0
    seller_score_items = data.find_all('span', class_ = 'mbg-l')
 
    for item in seller_score_items:
        
# Checking if seller score is available or not        
        if (re.search('\d+', str(item.text))):
            temp_item = int(re.findall('\d+', str(item.text))[0])
            score_count = score_count + 1
            if score_count < 2:
                Seller_Score.append(temp_item)            
#print(Seller_Score) 

 # Getting item price information
    Item_Price_items = data.find_all('span', id = 'prcIsum')
    if(Item_Price_items == []):
        Item_Price.append(None)
        
    for item in Item_Price_items:
        
# Checking if item price information is available or not        
        if (re.search('\d+.\d+', str(item.text))):
            temp_item = float(re.findall('\d+.\d+', str(item.text))[0])*100
            
            Item_Price.append(temp_item)  
            
        else:
            Item_Price.append(None)


#Getting number of items sold information

    temp_item = data.find('a', class_ = 'vi-txt-underline')
    if(temp_item != None):
        Num_Items_Sold.append(temp_item.text.split(' ')[0])
    else:
        Num_Items_Sold.append(None)
    
#print(Num_Items_Sold)

# Getting Best Offer items information    
    Best_Offer_items = data.find_all('div', class_ = 'vi-bbox-dspn u-flL w29 vi-price')
    if(Best_Offer_items == []):
        Best_Offer.append(None)
    for item in Best_Offer_items:
        if(item != None):
            Best_Offer.append(re.findall('\d+',str(item.text)))    


# Getting title information
    Title_items = data.find_all('h1', id = 'itemTitle')
    for item in Title_items:
# Using regex to extract only the title
        Title.append((item.text).split("\xa0")[1])
#print(Title)


# Getting information about return policy: It's either 'Returns Allowed' or 'No returns'

    Return_Allowed_items = data.find_all('span', class_ = 'w2b-sgl')
    #empty_string = empty_string.join(str(Return_Allowed_items.text))
  
    strin =''
    for i in range(0,len(Return_Allowed_items)):
        strin= strin+" "+Return_Allowed_items[i].text
    if('Returns' in strin):
        Return_Allowed.append(1)
    else:
        Return_Allowed.append(0)

# Getting shipping price information
    Shipping_Price_items = data.find_all('span', id = 'fshippingCost')
    if(Shipping_Price_items == []):
        Shipping_Price.append(None)

    for item in Shipping_Price_items:
        
#Using regex to extract only the digits of the price, or capture 'FREE' shipping or no shipping price information        
        if (re.search('\d+', str(item.text))):
            temp_item = float(re.findall('\d+.\d+', str(item.text))[0])*100
            Shipping_Price.append(temp_item)  
            
        elif (re.search('FREE', str(item.text))):
            Shipping_Price.append(0)


# Getting information about the condition of the item
    Condition_items = data.find_all('div', id = 'vi-itm-cond')
    for item in Condition_items:
        Condition.append(item.text) 
    eBay_item_type.append('sponsored')
print(len(Seller_Name), '\n', len(Seller_Score), '\n', len(Item_Price), '\n', len(Num_Items_Sold) , '\n', len(Best_Offer),'\n', len(Title),'\n',
      len(Return_Allowed),'\n', len(Shipping_Price),'\n', len(Condition),'\n',len(eBay_item_type))
```

```python
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  passwd=""
)
mycursor = mydb.cursor()

#SQL query for summary statistics of the eBay_items table columns
# MySQL automatically ignores NULL values while calculations, therefore there's not need to add an extra WHERE clause

sql_query = """ SELECT eBay_item_type, Condition_Item, count(Seller_Name), avg(Seller_Score) , min(Seller_Score), max(Seller_Score), stddev(Seller_Score),
                avg(Item_Price), min(Item_Price), max(Item_Price), stddev(Item_Price),
                avg(Num_Items_Sold), min(Num_Items_Sold), max(Num_Items_Sold), stddev(Num_Items_Sold),
                count(Title),
                avg(Shipping_Price), min(Shipping_Price), max(Shipping_Price), stddev(Shipping_Price)
                FROM ebay.ebay_items
                GROUP BY eBay_item_type, Condition_Item;"""

mycursor.execute(sql_query)
summary_stats = mycursor.fetchall()
# print(summary_stats)


df_summary_stats = pd.DataFrame(summary_stats, columns = ["eBay_item_type", "Condition_Item","Count Seller Name","Avg Seller Score","Min Seller Score",
                                                         "Max Seller Score","StdDev Seller Score","Avg Item Price",
                                                          "Min Item Price", "Max Item Price", "StdDev Item Price",
                                                          "Avg Number Items Sold","Min Number Items Sold","Max Number Items Sold",
                                                          "Std Dev Number Items Sold", "Count Title","Avg Shipping Price",
                                                          "Min Shipping Price", "Max Shipping Price","StdDev Shipping Price"])
print(df_summary_stats)
```
